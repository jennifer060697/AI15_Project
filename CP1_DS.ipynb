#### sigmoid 활성화 함수
- 이번 프로젝트에서는 이진 분류에 적합한 sigmoid 활성화 함수를 구현하여 활용한다.  
- sigmoid : $\sigma(x) = \frac{1}{1+e^{-x}}$

````python
# sigmoid 그래프 출력
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x) :
  return 1/(1+np.e**(-x))

x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)

fig = plt.figure(figsize = (8,5))
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.xlim(-10,10)
plt.title("Sigmoid", fontsize = 20)
plt.xlabel('x', fontsize = 15)
plt.ylabel('y', rotation = 0, fontsize = 15)
plt.yticks([0.0,0.5,1.0])
plt.axvline(0.0, color = 'k')
ax = plt.gca()
ax.yaxis.grid(True)

plt.show()
````

#### sigmoid 활성화 함수 장점
- 출력 값의 범위가 0~1 사이이며, 매끄러운 곡선을 가지므로, 역전파시 경사하강법을 시행할 때 기울기가 급격하게 변해서 발산하는 “기울기 폭주 (Gradient Exploding)" 가 발생하지 않는다.
- 분류는 0과 1로 나뉘며, 출력 값이 어디에 가까운지를 통해 어느 분류에 속하는지 분류한다.

#### sigmoid 활성화 함수 단점
- 출력값의 범위가 매우 좁기 때문에 경사하강법 수행시 0에 수렴하는 "기울기 소실 (Gradient Vanishing)" 이 발생할 수 있다.

### 손실 함수
- 손실 함수(Loss Function)는 지도학습(Supervised Learning) 시 알고리즘이 예측한 값과 실제 정답의 차이를 비교하기 위한 함수입니다.
- 즉, '학습 중에 알고리즘이 얼마나 잘못 예측하는 정도'를 확인하기 위한 함수로써 최적화(Optimization)를 위해 최소화하는 것이 목적인 함수입니다.
-  그래서 손실 함수를 목적 함수(Objective Function)라고도 부릅니다. 

#### 여러가지 손실 함수
- 평균 제곱 오차 (Mean Square Error, MSE)
$$\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat y_i)^{2}$$

- 평균 제곱근 오차 (Root Mean Square Error, RMSE)
$$\sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat y_i)^{2}}$$

- 이진 교차 엔트로피 오차 (Binary Cross-Entropy Error, BCEE)
$$-\frac{1}{n}\sum_{i=1}^{n}[y_i ln\hat y_i + (1-y_i)ln(1-\hat y_i)]$$

#### 이진 교차 엔트로피
- 교차 엔트로피 오차는 나누고자 하는 분류가 몇개인지에 따라 사용하는 손실함수가 바뀐다.
- 이는 사용되는 활성화 함수가 다르기 때문이다.
- 범주가 2개인 데이터는 시그모이드 함수 (sigmoid) 로 0\~1 걊을 반환하거나 하이퍼볼릭 탄젠트 함수 (Hyperbolic Tangent)를 사용하여 -1\~1 사이의 값을 반환한다. 활성화 함수 출력값이 단 하나의 스칼라 값을 가진다.
- 범주가 3개 이상이라면 총 합 1에 각 클래스에 속할 확률을 클래스 수만큼 반환하는 소프트맥스 함수 (Softmax)를 사용하여 클래스 수만큼 원소가 들어있는 배열을 반환하므로 평가 방법이 달라진다.
- 이진 교차 엔트로피 오차는 로그 손실 (Log Loss) 또는 로지스틱 손실 (Logistic Loss) 라고 불리며, 주로 로지스틱 회귀의 비용 함수로 사용된다.
- 이번 프로젝트에서는 이진 분류 손실값을 구하기 위해 이진 교차 엔트로피를 구현하여 활용한다.

### 가중치 초기화
- Xavier Initialization  
  - sigmoid, tanh 를 활성화 함수로 사용할 때 효과적이다.
  - 이전 은닉층의 노드 수에 맞추어 표준편차를 변화시킨다.  
  - 이전 은닉층 노드 개수 n개, 현재 은닉층 노드 개수 m개 일 때,  
표준편차 = $\frac{2}{\sqrt{n+m}}$ 로 하는 정규분포로 가중치를 랜덤 초기화한다.
- He Initialization.  
  - ReLU 함수를 활성화 함수로 사용할 때 효과적이다.  
  - 이전 은닉층 노드 개수 n개 일 때,  
표준편차 = $\sqrt{\frac{2}{n}}$ 로 하는 정규분포로 가중치를 랜덤 초기화한다.

- 이 프로젝트에서는 sigmoid 를 활성화 함수로 사용하기 때문에 사비에르 초기화를 사용하여 가중치를 초기화한다.

### 성능 평가 지표 (분류)
- Precision (정밀도) : $\frac{TP}{TP + FP}$
- Recall (재현율) : $\frac{TP}{TP + FN}$
- Accuracy (정확도) : $\frac{TP + TN}{TP + FN + FP + TN}$
- F1 score : precision, recall 조화 평균

````python
def data_save(url = "https://drive.google.com/uc?export=download&id=1SCO0ZGL_EDGWc9Le0JFDw9eww86xk1xJ", file_name = "datafile.csv") :
  """
  데이터 로컬에 저장
  param : url (데이터 링크), file_name (로컬에 저장될 파일명)
  return : 없음
  """
  try:
    # 링크에서 읽어오기
    r = requests.get(url).content
  except:
    print("Invalid URL or some error occured while making the GET request to the specified URL")
    return
  # 읽어온 데이터 로컬에 저장
  f = open(file_name, "wb")
  f.write(r)
  f.close()

def data_load(file_path = "/content/datafile.csv") :
  """
  로컬에 저장한 데이터 로드
  param : file_path (로컬에서 로드할 파일 경로)
  return : 판다스 데이터프레임
  """
  # 1. 코랩에 드라이브 마운트로 불러오기
  # from google.colab import drive
  # drive.mount('/content/drive')
  # location = "/content/drive/MyDrive/binary_dataset.csv"
  # dataset = pd.read_csv(location)
  # return dataset

  # 2. 링크 데이터 바로 불러오기
  # rawData = pd.read_csv(io.StringIO(r.decode('utf-8')))

  # 3. 로컬에 저장한 데이터 불러오기
  return pd.read_csv(file_path)

def initialization_parameter(input_dimension, my_dimension) :
  """                                                                                                                                                                                                                                                             
  사비에르 초기화로 파라미터, 편향 생성
  param : input_dimension (이전층 노드 개수), my_dimension (현재층 노드 개수)
  return : 초기화된 weights, bias 행렬
  """
  # Xavier initialization
  # 평균 = 0, 표준편차 = 2/sqrt(n+m), (n = 이전층 노드 개수, m = 현재층 노드 개수) 정규분포에서 랜덤 선택
  weights = np.random.normal(0, 2/((input_dimension+my_dimension)**0.5), size=[input_dimension, my_dimension])
  #bias = np.zeros([my_dimension]) # bias 0으로 초기화
  bias = np.random.normal(0, 2/((input_dimension+my_dimension)**0.5), size = [my_dimension]) # 가중치와 동일한 사비에르 초기화를 편향의 기본값으로 주었음
  return weights, bias

def data_randmix(df) :
  """
  데이터 뒤섞기
  데이터를 학습하는 과정에서 해당 데이터의 순서에 의존하지 않고 임의의 데이터가 입력되는 것으로 간주하기 위함이다.
  param : df (데이터 프레임)
  return : 섞인 데이터
  """
  return df.sample(frac=1).reset_index(drop=True)

def split_target_data(df) :
  """
  X,y 데이터 분리
  주어진 데이터 셋에 맞춘 함수
  param : df (데이터 프레임)
  return : X, y
  """
  return df.iloc[:,range(8)], df['y']

def split_train_test_data(df, test_ratio) :
  """
  셔플 없이 비율대로 앞뒤로 잘라서 train, test 데이터 분리
  param : df (데이터프레임), test_ratio (테스트셋 비율)
  return : train, test
  """
  n = int(len(df)*(1-test_ratio))
  return df[:n], df[n:]

def normalize(df) :
  """
  minmax 정규화
  param : df (데이터프레임)
  return : df (정규화된 데이터프레임)
  """
  df = (df - df.min())/(df.max() - df.min())
  return df


def mini_batch(df, size) :
  """
  미니배치
  인덱스 넘버로 미니배치를 구현. 에포크마다 랜덤 추출 되도록 인덱스를 섞어서 제너레이터로 데이터 반환
  param : df (데이터프레임), size (미니배치 크기)
  return : 미니배치 데이터프레임 제너레이터
  """
  index_list = list(df.index)
  random.shuffle(index_list)
  for i in range(0,len(index_list),size) :
    ind = index_list[i:i+size]
    yield split_target_data(df.loc[ind])

def sigmoid(x) :
  """
  시그모이드
  param : x
  return : 시그모이드 통과값
  """
  return 1/(1+np.e**(-x))

def forward_prop(input_data, weights, bias) :
  """
  데이터 하나가 들어왔을 때 신경망 한층 연산
  param : input_data (단일 데이터), weights, bias
  return : 연산 결과 np 행렬
  """
  cal = input_data.dot(weights) + bias # 순전파 행렬 연산
  sig_cal = sigmoid(cal) # 활성화함수 : 시그모이드
  return np.array(sig_cal)

def cal_lost(sig_cal, y) :
  """
  배치별 손실(이진 교차 엔트로피)값 연산
  param : sig_cal (배치 1개의 모델 최종 output 행렬), y (배치 1개의 타겟값)
  return : 배치의 평균 이진 교차 엔트로피
  """
  return -sum([y*np.log(p)+(1-y)*np.log(1-p) for p,y in zip(sig_cal, y)]) / len(y) # 평균 이진 교차 엔트로피 연산

def accuracy(sig_cal, y, threshold = 0.5) :
  """
  배치별 평균 정확도 연산
  param : sig_cal (배치 1개의 모델 최종 output 행렬), y (배치 1개의 타겟값), threshold (임계값)
  return : 배치의 평균 정확도
  """
  pred = [1 if i>=threshold else 0 for i in sig_cal]
  acc = [1 if p==y else 0 for p,y in zip(pred, y)]
  return sum(acc)/len(acc)

class model :
  """
  모델 클래스
  param : input_dim (input 데이터의 차원)
  """
  def __init__(self, input_dim) :
    """
    내장 변수
    weights : 레이어들 가중치
    bias : 레이어들 편향
    dim : 레이어들 차원
    fin_dim : 출력 차원
    forward_prop_by_batch : 배치별 순전파 결과값
    loss_by_batch : 배치별 loss 값
    acc_by_batch : 배치별 accuracy 값
    """
    self.weights = [] # 레이어들 가중치
    self.bias = [] # 레이어들 편향
    self.dim = [input_dim] # 레이어들 차원
    self.fin_dim = self.dim[0] # 마지막 차원

    self.forward_prop_by_batch = [] # 배치별 순전파 결과값
    self.loss_by_batch = [] # 배치별 loss 값
    self.acc_by_batch = [] # 배치별 accuracy 값

  def add_dense_layor(self, out_dim) :
    """
    모델의 맨 끝에 out_dim 차원의 dense layor 를 추가한다. 초기화된 파라미터 행렬을 저장한다.
    param : out_dim (추가하려는 dense layor 의 차원)
    return : 없음
    """
    in_dim = self.fin_dim # 현재 모델의 마지막 차원을 input 차원으로 설정
    w,b = initialization_parameter(in_dim, out_dim) # 파라미터 초기화
    self.weights.append(w) # 파라미터 저장
    self.bias.append(b)
    self.dim.append(out_dim) # 차원값 저장
    self.fin_dim = out_dim # 현재 모델의 마지막 차원값 재설정

  def forward_propagation(self, batch) :
    """
    미니배치를 받아서 순전파 진행, 배치별 순전파 결과와 loss, accuracy 저장
    param : batch (미니배치 제너레이터)
    return : 없음
    """
    # 초기화
    self.forward_prop_by_batch = []
    self.loss_by_batch = []
    self.acc_by_batch = []

    # 배치별 순전파 연산 및 loss, acc 연산
    for X,y in batch :
      X = np.array(X)
      for w, b in zip(self.weights, self.bias) :
        X = forward_prop(X, w, b)
      self.forward_prop_by_batch.append(X)
      self.loss_by_batch.append(cal_lost(X, y))
      self.acc_by_batch.append(accuracy(X, y))




````

````python
def main() :
  # 데이터 다운하여 저장 및 로드
  data_save(url = "https://drive.google.com/uc?export=download&id=1SCO0ZGL_EDGWc9Le0JFDw9eww86xk1xJ", file_name = "datafile.csv")
  df = data_load(file_path = "/content/datafile.csv")

  # 데이터 섞기
  df = data_randmix(df)

  # train, test 데이터 분리
  train, test = split_train_test_data(df, 0.2)

  # 모델 만들기
  M = model(8) # input dimension 을 파라미터로 넘기면서 시작 = input layer
  M.add_dense_layor(5) # (노드 5개) 밀집 레이어 (+ 시그모이드) 추가 = hidden layer
  M.add_dense_layor(10) # (노드 10개) 밀집 레이어 (+ 시그모이드) 추가 = hidden layer
  M.add_dense_layor(1) # (노드 1개) 밀집 레이어 (+ 시그모이드) 추가 = output layer
  
  # train data 에 대해 에포크 1회만 진행
  for i in range(1,2) :

    # 미니배치 + X,y 데이터 분리
    train_batch = mini_batch(train, 4)
    #test_batch = mini_batch(test, 4)

    # 순전파 진행
    M.forward_propagation(train_batch)
    
    # loss, accuracy 구하기
    loss = np.mean(M.loss_by_batch)
    accuracy = np.mean(M.acc_by_batch)
    
    print(f"[Epoch {i}] TrainData - Loss = {loss:.3f}, Accuracy = {accuracy:.3f}")
````

````python
main()
````

## 역전파 이론
역전파(Backpropagation)는 먼저 계산 결과와 정답의 오차를 구해 이 오차에 관여하는 값들의 가증치를 수정하여 오차가 작아지는 방향으로 일정 횟수를 반복해 수정하는 방법이다. 오차역전파 또는 오류역전파라고도 불린다.



### 학습 과정
1. 임의의 초기 가중치($W$)를 준 뒤 결과 ($Y_{out}$)를 계산한다.  
2. 계산 결과와 우리가 원하는 값 사이의 오차를 구한다.  
3. 경사 하강법을 이용해 바로 앞 가중치를 오차가 작아지는 방향으로 업데이트한다.  
4. 위 과정을 더는 오차가 줄어들지 않을 때까지 반복한다.  

### 경사하강법
오차가 작아지는 방향으로 업데이트한다'는 의미는 미분 값이 0에 가까워지는 방향으로 나아간다는 말이다. 즉, '기울기가 0이 되는 방향'으로 나아가야 하는데, 이 말은 가중치에서 기울기를 뺐을 때 가중치의 변화가 전혀 없는 상태를 말한다.
적절한 값으로 값을 갱신하기 위해 학습률(α) 값을 조정한다.
  $$W(t+1)=W(t)-α\frac{\partial{Loss}}{\partial{W}}$$

경사하강법을 통한 최소값 찾기는 local minimum 문제에 빠질 수 있다.

### Chain Rule
$$\frac{dy}{dx} = \frac{dy}{du}\cdot\frac{du}{dx}$$
체인룰을 통해 $\frac{\partial{Loss}}{\partial{W}}$ 값을 구할 수 있다.

### 역전파 모식도

dense 레이어의 역전파 학습은 현재 층의 input 값, 다음 층에서 받아올 수 있는 역전파 연산값의 행렬곱으로 표현할 수 있다.

## 역전파 구현

````python
class dense :
  """
  dense layer class
  """
  def __init__(self, in_dim:int, out_dim:int) :
    """
    param
      in_dim (int) : input dimension
      out_dim (int) : output dimension
    """
    self.W , self.b = initialization_parameter(in_dim, out_dim)
    self.dW = None # 가중치 변화량
    self.db = None # 편향 변화량
    self.x = None # input data
  
  def forward(self, x:np.array) -> np.array :
    """
    dense 레이어의 순전파
    param
      x (np.array) : input data
    return
      out (np.array) : 순전파 결과값
    """
    self.x = x
    out = np.dot(x,self.W) + self.b
    return out

  def backward(self, dout:np.array, learning_rate:float) -> np.array : # dout : 직전 역전파 진행에서 받아온 미분값
    """
    dense 레이어의 역전파
    param
      dout (np.array) : 이전 역전파에서 받아온 값
      learning_rate (float) : 학습률
    return
      dx (np.array) : 다음 역전파에 넘길 값
    """
    dx = np.dot(dout,self.W.T) # 다음 역전파층에서 사용할 미분값
    self.dW = np.dot(self.x.T, dout) # 가중치 변화량 계산
    self.db = np.sum(dout, axis = 0) # 편향 변화량 계산
    self.W -= learning_rate * self.dW # 가중치 업데이트
    self.b -= learning_rate * self.db # 편향 업데이트

    return dx

class sigmoid :
  """
  sigmoid layer
  """
  def __init__(self) :
    self.out = None # 시그모이드 통과 결과값

  def forward(self, x:np.array)->np.array :
    """
    시그모이드 순전파
    param
      x (np.array) : input
    return
      out (np.array) : sigmoid(x)
    """
    out = 1/(1+np.exp(-x))
    self.out = out
    return out
  
  def backward(self, dout:np.array, _)->np.array :
    """
    시그모이드 역전파
    param
      dout (np.array) : 이전 역전파에서 받아온 값
    return
      dx (np.array) : 다음 역전파에 넘길 값
    """
    dx = dout*(1.0-self.out)*self.out
    return dx

class relu :
  """
  relu layer
  """
  def __init__(self) :
    self.mask = None

  def forward(self, x:np.array)->np.array :
    """
    relu 순전파
    param
      x (np.array) : input
    return
      out (np.array) : relu(x)
    """
    self.mask = (x <= 0)
    out = x.copy()
    out[self.mask] = 0
    return out

  def backward(self, dout:np.array, _)->np.array :
    """
    relu 역전파
    param
      dout (np.array) : 이전 역전파에서 받아온 값
    return
      dx (np.array) : 다음 역전파에 넘길 값
    """
    dout[self.mask] = 0
    dx = dout
    return dx
  

class model_with_back_propagation :
  """
  역전파 학습 구현 모델
  """
  def __init__(self, input_dim:int) :
    self.layers_info = [('input',input_dim)] # 레이어 정보
    self.fin_dim = input_dim # 마지막 차원
    self.layers = [] # 레이어 배열
    self.y = None # 예측 결과값
    self.t = None # 실제 타겟값
    self.loss = None # 로스값
  
  def add_dense_layer(self, out_dim:int) :
    """
    모델에 dense layer 추가
    param
      out_dim (int) : 레이어 차원
    """
    self.layers.append(dense(self.fin_dim, out_dim))
    print(f"dense layer added : input dim = {self.fin_dim}, output dim = {out_dim}")
    self.layers_info.append(('dense',out_dim))
    self.fin_dim = out_dim

  def add_sigmoid_layer(self) :
    """
    모델에 sigmoid layer 추가
    """
    self.layers.append(sigmoid())
    print(f"sigmoid layer added")
    self.layers_info.append(('sigmoid',self.fin_dim))

  def add_relu_layer(self) :
    """
    모델에 relu layer 추가
    """
    self.layers.append(relu())
    print(f"relu layer added")
    self.layers_info.append(('relu',self.fin_dim))
  
  def forward_pass(self, x:np.array) :
    """
    모델 순전파 진행
    param
     x (np.array) : input data
    return
     x (np.array) : 순전파 결과값
    """
    for layer in self.layers :
      x = layer.forward(x)
    self.y = x
    return x
  
  def find_loss(self, t:np.array)->float :
    """
    이진 바이너리 교차 엔트로피 loss 값 계산
    param
      t (np.array) : 실제 타겟값
    return
      self.loss (float) : 평균 loss 계산값
    """
    if not self.y.all() :
      print("Error : need forward_pass")
      return
    self.loss = np.squeeze(cal_lost(self.y, t))
    self.t = t
    return self.loss

  def acc(self, t:np.array)->np.array :
    """
    정확도 계산
    param
      t (np.array) : 실제 타겟값
    return
      ac : 정확도 연산값
    """
    if not self.y.all() :
      print("Error : need forward_pass")
      return
    ac = accuracy(self.y,t)
    return ac

  def back_prop(self, t:np.array, learning_rate:float) :
    """
    역전파와 학습
    param
      t (np.array) : 실제 타겟값
      learning_rate (float) : 학습률
    """
    t = np.array(t).reshape(-1,1)

    dout = -(t/self.y - (1-t)/(1-self.y)) # 손실함수 미분 역전파 계산

    for layer in self.layers[::-1] : # 레이어 역순으로 역전파와 학습 진행
      dout = layer.backward(dout, learning_rate)

````

````python
def main() :

  # 데이터 다운하여 저장 및 로드
  data_save(url = "https://drive.google.com/uc?export=download&id=1SCO0ZGL_EDGWc9Le0JFDw9eww86xk1xJ", file_name = "datafile.csv")
  df = data_load(file_path = "/content/datafile.csv")

  # 데이터 섞기
  df = data_randmix(df)

  # train, test 데이터 분리
  train, test = split_train_test_data(df, 0.2)

  # split
  X_train, y_train = split_target_data(train)
  X_test, y_test = split_target_data(test)

  # 데이터 정규화
  X_train_norm = normalize(X_train)
  X_test_norm = normalize(X_test)
  train_norm = pd.concat([X_train_norm,y_train], axis=1)

  # 모델 만들기
  M = model_with_back_propagation(8)
  M.add_dense_layer(20)
  M.add_relu_layer()
  M.add_dense_layer(1)
  M.add_sigmoid_layer()

  print()

  # 역전파 및 학습
  n = 1000 # 에포크 횟수
  batch_size = 4
  for i in range(1,n+1) : # 에포크 n회
    # 미니배치 만들기
    train_batch = mini_batch(train_norm, batch_size)
    loss = 0
    acc = 0
    # 배치별 학습
    for x,y in train_batch :
      M.forward_pass(x)
      if i%100 == 0 :
        loss += M.find_loss(y)
        acc += M.acc(y)
      else : M.find_loss(y)
      M.back_prop(y, 0.01)
    # 출력
    if i%100 == 0 :
      print(f"[Epoch {i}] TrainData - Loss = {loss/batch_size:.5f}, Accuracy = {acc/batch_size}")

  print()

  print("-----check test set-----")
  M.forward_pass(X_test_norm)
  print(f"test accuracy : {M.acc(y_test)}")
  pred = ((M.y >= 0.5).reshape(-1) == np.array(y_test))
  for i in pred :
    if i : print("correct")
    else : print("wrong")
````
