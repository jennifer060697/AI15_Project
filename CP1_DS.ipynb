### ÌôúÏÑ±Ìôî Ìï®Ïàò

linearÌïú Ïó∞ÏÇ∞ÏùÑ Í∞ñÎäî layerÎ•º ÏàòÏã≠Í∞ú ÏåìÏïÑÎèÑ Í≤∞Íµ≠ Ïù¥Îäî ÌïòÎÇòÏùò linear Ïó∞ÏÇ∞ÏúºÎ°ú ÎÇòÌÉÄÎÇº Ïàò ÏûàÎã§.  
ÌôúÏÑ±Ìôî Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÎ©¥ ÏûÖÎ†•Í∞íÏóê ÎåÄÌïú Ï∂úÎ†•Í∞íÏù¥ linearÌïòÍ≤å ÎÇòÏò§ÏßÄ ÏïäÏúºÎØÄÎ°ú ÏÑ†ÌòïÎ∂ÑÎ•òÍ∏∞Î•º ÎπÑÏÑ†Ìòï ÏãúÏä§ÌÖúÏúºÎ°ú ÎßåÎì§ Ïàò ÏûàÎã§.
Îî∞ÎùºÏÑú MLP(Multiple layer perceptron)Îäî Îã®ÏßÄ linear layerÎ•º Ïó¨Îü¨Í∞ú ÏåìÎäî Í∞úÎÖêÏù¥ ÏïÑÎãå ÌôúÏÑ±Ìôî Ìï®ÏàòÎ•º Ïù¥Ïö©Ìïú non-linear ÏãúÏä§ÌÖúÏùÑ Ïó¨Îü¨ layerÎ°ú ÏåìÎäî Í∞úÎÖêÏù¥Îã§.

#### sigmoid ÌôúÏÑ±Ìôî Ìï®Ïàò
- Ïù¥Î≤à ÌîÑÎ°úÏ†ùÌä∏ÏóêÏÑúÎäî Ïù¥ÏßÑ Î∂ÑÎ•òÏóê Ï†ÅÌï©Ìïú sigmoid ÌôúÏÑ±Ìôî Ìï®ÏàòÎ•º Íµ¨ÌòÑÌïòÏó¨ ÌôúÏö©ÌïúÎã§.  
- sigmoid : $\sigma(x) = \frac{1}{1+e^{-x}}$

````python
# sigmoid Í∑∏ÎûòÌîÑ Ï∂úÎ†•
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x) :
  return 1/(1+np.e**(-x))

x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)

fig = plt.figure(figsize = (8,5))
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.xlim(-10,10)
plt.title("Sigmoid", fontsize = 20)
plt.xlabel('x', fontsize = 15)
plt.ylabel('y', rotation = 0, fontsize = 15)
plt.yticks([0.0,0.5,1.0])
plt.axvline(0.0, color = 'k')
ax = plt.gca()
ax.yaxis.grid(True)

plt.show()
````

#### sigmoid ÌôúÏÑ±Ìôî Ìï®Ïàò Ïû•Ï†ê
- Ï∂úÎ†• Í∞íÏùò Î≤îÏúÑÍ∞Ä 0~1 ÏÇ¨Ïù¥Ïù¥Î©∞, Îß§ÎÅÑÎü¨Ïö¥ Í≥°ÏÑ†ÏùÑ Í∞ÄÏßÄÎØÄÎ°ú, Ïó≠Ï†ÑÌååÏãú Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏùÑ ÏãúÌñâÌï† Îïå Í∏∞Ïö∏Í∏∞Í∞Ä Í∏âÍ≤©ÌïòÍ≤å Î≥ÄÌï¥ÏÑú Î∞úÏÇ∞ÌïòÎäî ‚ÄúÍ∏∞Ïö∏Í∏∞ Ìè≠Ï£º (Gradient Exploding)" Í∞Ä Î∞úÏÉùÌïòÏßÄ ÏïäÎäîÎã§.
- Î∂ÑÎ•òÎäî 0Í≥º 1Î°ú ÎÇòÎâòÎ©∞, Ï∂úÎ†• Í∞íÏù¥ Ïñ¥ÎîîÏóê Í∞ÄÍπåÏö¥ÏßÄÎ•º ÌÜµÌï¥ Ïñ¥Îäê Î∂ÑÎ•òÏóê ÏÜçÌïòÎäîÏßÄ Î∂ÑÎ•òÌïúÎã§.

#### sigmoid ÌôúÏÑ±Ìôî Ìï®Ïàò Îã®Ï†ê

### ÏÜêÏã§ Ìï®Ïàò
- ÏÜêÏã§ Ìï®Ïàò(Loss Function)Îäî ÏßÄÎèÑÌïôÏäµ(Supervised Learning) Ïãú ÏïåÍ≥†Î¶¨Ï¶òÏù¥ ÏòàÏ∏°Ìïú Í∞íÍ≥º Ïã§Ï†ú Ï†ïÎãµÏùò Ï∞®Ïù¥Î•º ÎπÑÍµêÌïòÍ∏∞ ÏúÑÌïú Ìï®ÏàòÏûÖÎãàÎã§.
- Ï¶â, 'ÌïôÏäµ Ï§ëÏóê ÏïåÍ≥†Î¶¨Ï¶òÏù¥ ÏñºÎßàÎÇò ÏûòÎ™ª ÏòàÏ∏°ÌïòÎäî Ï†ïÎèÑ'Î•º ÌôïÏù∏ÌïòÍ∏∞ ÏúÑÌïú Ìï®ÏàòÎ°úÏç® ÏµúÏ†ÅÌôî(Optimization)Î•º ÏúÑÌï¥ ÏµúÏÜåÌôîÌïòÎäî Í≤ÉÏù¥ Î™©Ï†ÅÏù∏ Ìï®ÏàòÏûÖÎãàÎã§.
-  Í∑∏ÎûòÏÑú ÏÜêÏã§ Ìï®ÏàòÎ•º Î™©Ï†Å Ìï®Ïàò(Objective Function)ÎùºÍ≥†ÎèÑ Î∂ÄÎ¶ÖÎãàÎã§. 

#### Ïó¨Îü¨Í∞ÄÏßÄ ÏÜêÏã§ Ìï®Ïàò
- ÌèâÍ∑† Ï†úÍ≥± Ïò§Ï∞® (Mean Square Error, MSE)
$$\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat y_i)^{2}$$

- ÌèâÍ∑† Ï†úÍ≥±Í∑º Ïò§Ï∞® (Root Mean Square Error, RMSE)
$$\sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat y_i)^{2}}$$

- Ïù¥ÏßÑ ÍµêÏ∞® ÏóîÌä∏Î°úÌîº Ïò§Ï∞® (Binary Cross-Entropy Error, BCEE)
$$-\frac{1}{n}\sum_{i=1}^{n}[y_i ln\hat y_i + (1-y_i)ln(1-\hat y_i)]$$

#### Ïù¥ÏßÑ ÍµêÏ∞® ÏóîÌä∏Î°úÌîº
- ÍµêÏ∞® ÏóîÌä∏Î°úÌîº Ïò§Ï∞®Îäî ÎÇòÎàÑÍ≥†Ïûê ÌïòÎäî Î∂ÑÎ•òÍ∞Ä Î™áÍ∞úÏù∏ÏßÄÏóê Îî∞Îùº ÏÇ¨Ïö©ÌïòÎäî ÏÜêÏã§Ìï®ÏàòÍ∞Ä Î∞îÎÄêÎã§.
- Ïù¥Îäî ÏÇ¨Ïö©ÎêòÎäî ÌôúÏÑ±Ìôî Ìï®ÏàòÍ∞Ä Îã§Î•¥Í∏∞ ÎïåÎ¨∏Ïù¥Îã§.
- Î≤îÏ£ºÍ∞Ä 2Í∞úÏù∏ Îç∞Ïù¥ÌÑ∞Îäî ÏãúÍ∑∏Î™®Ïù¥Îìú Ìï®Ïàò (sigmoid) Î°ú 0\~1 Í±äÏùÑ Î∞òÌôòÌïòÍ±∞ÎÇò ÌïòÏù¥ÌçºÎ≥ºÎ¶≠ ÌÉÑÏ††Ìä∏ Ìï®Ïàò (Hyperbolic Tangent)Î•º ÏÇ¨Ïö©ÌïòÏó¨ -1\~1 ÏÇ¨Ïù¥Ïùò Í∞íÏùÑ Î∞òÌôòÌïúÎã§. ÌôúÏÑ±Ìôî Ìï®Ïàò Ï∂úÎ†•Í∞íÏù¥ Îã® ÌïòÎÇòÏùò Ïä§ÏπºÎùº Í∞íÏùÑ Í∞ÄÏßÑÎã§.
- Î≤îÏ£ºÍ∞Ä 3Í∞ú Ïù¥ÏÉÅÏù¥ÎùºÎ©¥ Ï¥ù Ìï© 1Ïóê Í∞Å ÌÅ¥ÎûòÏä§Ïóê ÏÜçÌï† ÌôïÎ•†ÏùÑ ÌÅ¥ÎûòÏä§ ÏàòÎßåÌÅº Î∞òÌôòÌïòÎäî ÏÜåÌîÑÌä∏Îß•Ïä§ Ìï®Ïàò (Softmax)Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÌÅ¥ÎûòÏä§ ÏàòÎßåÌÅº ÏõêÏÜåÍ∞Ä Îì§Ïñ¥ÏûàÎäî Î∞∞Ïó¥ÏùÑ Î∞òÌôòÌïòÎØÄÎ°ú ÌèâÍ∞Ä Î∞©Î≤ïÏù¥ Îã¨ÎùºÏßÑÎã§.
- Ïù¥ÏßÑ ÍµêÏ∞® ÏóîÌä∏Î°úÌîº Ïò§Ï∞®Îäî Î°úÍ∑∏ ÏÜêÏã§ (Log Loss) ÎòêÎäî Î°úÏßÄÏä§Ìã± ÏÜêÏã§ (Logistic Loss) ÎùºÍ≥† Î∂àÎ¶¨Î©∞, Ï£ºÎ°ú Î°úÏßÄÏä§Ìã± ÌöåÍ∑ÄÏùò ÎπÑÏö© Ìï®ÏàòÎ°ú ÏÇ¨Ïö©ÎêúÎã§.
- Ïù¥Î≤à ÌîÑÎ°úÏ†ùÌä∏ÏóêÏÑúÎäî Ïù¥ÏßÑ Î∂ÑÎ•ò ÏÜêÏã§Í∞íÏùÑ Íµ¨ÌïòÍ∏∞ ÏúÑÌï¥ Ïù¥ÏßÑ ÍµêÏ∞® ÏóîÌä∏Î°úÌîºÎ•º Íµ¨ÌòÑÌïòÏó¨ ÌôúÏö©ÌïúÎã§.

  - sigmoid, tanh Î•º ÌôúÏÑ±Ìôî Ìï®ÏàòÎ°ú ÏÇ¨Ïö©Ìï† Îïå Ìö®Í≥ºÏ†ÅÏù¥Îã§.
  - Ïù¥Ï†Ñ ÏùÄÎãâÏ∏µÏùò ÎÖ∏Îìú ÏàòÏóê ÎßûÏ∂îÏñ¥ ÌëúÏ§ÄÌé∏Ï∞®Î•º Î≥ÄÌôîÏãúÌÇ®Îã§.  
  - Ïù¥Ï†Ñ ÏùÄÎãâÏ∏µ ÎÖ∏Îìú Í∞úÏàò nÍ∞ú, ÌòÑÏû¨ ÏùÄÎãâÏ∏µ ÎÖ∏Îìú Í∞úÏàò mÍ∞ú Ïùº Îïå,  
ÌëúÏ§ÄÌé∏Ï∞® = $\frac{2}{\sqrt{n+m}}$ Î°ú ÌïòÎäî Ï†ïÍ∑úÎ∂ÑÌè¨Î°ú Í∞ÄÏ§ëÏπòÎ•º ÎûúÎç§ Ï¥àÍ∏∞ÌôîÌïúÎã§.
- He Initialization.  
  - ReLU Ìï®ÏàòÎ•º ÌôúÏÑ±Ìôî Ìï®ÏàòÎ°ú ÏÇ¨Ïö©Ìï† Îïå Ìö®Í≥ºÏ†ÅÏù¥Îã§.  
  - Ïù¥Ï†Ñ ÏùÄÎãâÏ∏µ ÎÖ∏Îìú Í∞úÏàò nÍ∞ú Ïùº Îïå,  
ÌëúÏ§ÄÌé∏Ï∞® = $\sqrt{\frac{2}{n}}$ Î°ú ÌïòÎäî Ï†ïÍ∑úÎ∂ÑÌè¨Î°ú Í∞ÄÏ§ëÏπòÎ•º ÎûúÎç§ Ï¥àÍ∏∞ÌôîÌïúÎã§.

- Ïù¥ ÌîÑÎ°úÏ†ùÌä∏ÏóêÏÑúÎäî sigmoid Î•º ÌôúÏÑ±Ìôî Ìï®ÏàòÎ°ú ÏÇ¨Ïö©ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê ÏÇ¨ÎπÑÏóêÎ•¥ Ï¥àÍ∏∞ÌôîÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Í∞ÄÏ§ëÏπòÎ•º Ï¥àÍ∏∞ÌôîÌïúÎã§.

### ÏÑ±Îä• ÌèâÍ∞Ä ÏßÄÌëú (Î∂ÑÎ•ò)
- Precision (Ï†ïÎ∞ÄÎèÑ) : $\frac{TP}{TP + FP}$
- Recall (Ïû¨ÌòÑÏú®) : $\frac{TP}{TP + FN}$
- Accuracy (Ï†ïÌôïÎèÑ) : $\frac{TP + TN}{TP + FN + FP + TN}$
- F1 score : precision, recall Ï°∞Ìôî ÌèâÍ∑†

````python
# ÏÇ¨Ïö© Í∞ÄÎä•Ìïú ÎùºÏù¥Î∏åÎü¨Î¶¨
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import random
import requests
import io
````

````python
def data_save(url = "https://drive.google.com/uc?export=download&id=1SCO0ZGL_EDGWc9Le0JFDw9eww86xk1xJ", file_name = "datafile.csv") :
  """
  Îç∞Ïù¥ÌÑ∞ Î°úÏª¨Ïóê Ï†ÄÏû•
  param : url (Îç∞Ïù¥ÌÑ∞ ÎßÅÌÅ¨), file_name (Î°úÏª¨Ïóê Ï†ÄÏû•Îê† ÌååÏùºÎ™Ö)
  return : ÏóÜÏùå
  """
  try:
    # ÎßÅÌÅ¨ÏóêÏÑú ÏùΩÏñ¥Ïò§Í∏∞
    r = requests.get(url).content
  except:
    print("Invalid URL or some error occured while making the GET request to the specified URL")
    return
  # ÏùΩÏñ¥Ïò® Îç∞Ïù¥ÌÑ∞ Î°úÏª¨Ïóê Ï†ÄÏû•
  f = open(file_name, "wb")
  f.write(r)
  f.close()

def data_load(file_path = "/content/datafile.csv") :
  """
  Î°úÏª¨Ïóê Ï†ÄÏû•Ìïú Îç∞Ïù¥ÌÑ∞ Î°úÎìú
  param : file_path (Î°úÏª¨ÏóêÏÑú Î°úÎìúÌï† ÌååÏùº Í≤ΩÎ°ú)
  return : ÌåêÎã§Ïä§ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ
  """
  # 1. ÏΩîÎû©Ïóê ÎìúÎùºÏù¥Î∏å ÎßàÏö¥Ìä∏Î°ú Î∂àÎü¨Ïò§Í∏∞
  # from google.colab import drive
  # drive.mount('/content/drive')
  # location = "/content/drive/MyDrive/binary_dataset.csv"
  # dataset = pd.read_csv(location)
  # return dataset

  # 2. ÎßÅÌÅ¨ Îç∞Ïù¥ÌÑ∞ Î∞îÎ°ú Î∂àÎü¨Ïò§Í∏∞
  # rawData = pd.read_csv(io.StringIO(r.decode('utf-8')))

  # 3. Î°úÏª¨Ïóê Ï†ÄÏû•Ìïú Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
  return pd.read_csv(file_path)

def initialization_parameter(input_dimension, my_dimension) :
  """                                                                                                                                                                                                                                                             
  ÏÇ¨ÎπÑÏóêÎ•¥ Ï¥àÍ∏∞ÌôîÎ°ú ÌååÎùºÎØ∏ÌÑ∞, Ìé∏Ìñ• ÏÉùÏÑ±
  param : input_dimension (Ïù¥Ï†ÑÏ∏µ ÎÖ∏Îìú Í∞úÏàò), my_dimension (ÌòÑÏû¨Ï∏µ ÎÖ∏Îìú Í∞úÏàò)
  return : Ï¥àÍ∏∞ÌôîÎêú weights, bias ÌñâÎ†¨
  """
  # Xavier initialization
  # ÌèâÍ∑† = 0, ÌëúÏ§ÄÌé∏Ï∞® = 2/sqrt(n+m), (n = Ïù¥Ï†ÑÏ∏µ ÎÖ∏Îìú Í∞úÏàò, m = ÌòÑÏû¨Ï∏µ ÎÖ∏Îìú Í∞úÏàò) Ï†ïÍ∑úÎ∂ÑÌè¨ÏóêÏÑú ÎûúÎç§ ÏÑ†ÌÉù
  weights = np.random.normal(0, 2/((input_dimension+my_dimension)**0.5), size=[input_dimension, my_dimension])
  #bias = np.zeros([my_dimension]) # bias 0ÏúºÎ°ú Ï¥àÍ∏∞Ìôî
  bias = np.random.normal(0, 2/((input_dimension+my_dimension)**0.5), size = [my_dimension]) # Í∞ÄÏ§ëÏπòÏôÄ ÎèôÏùºÌïú ÏÇ¨ÎπÑÏóêÎ•¥ Ï¥àÍ∏∞ÌôîÎ•º Ìé∏Ìñ•Ïùò Í∏∞Î≥∏Í∞íÏúºÎ°ú Ï£ºÏóàÏùå
  return weights, bias

def data_randmix(df) :
  """
  Îç∞Ïù¥ÌÑ∞ Îí§ÏÑûÍ∏∞
  Îç∞Ïù¥ÌÑ∞Î•º ÌïôÏäµÌïòÎäî Í≥ºÏ†ïÏóêÏÑú Ìï¥Îãπ Îç∞Ïù¥ÌÑ∞Ïùò ÏàúÏÑúÏóê ÏùòÏ°¥ÌïòÏßÄ ÏïäÍ≥† ÏûÑÏùòÏùò Îç∞Ïù¥ÌÑ∞Í∞Ä ÏûÖÎ†•ÎêòÎäî Í≤ÉÏúºÎ°ú Í∞ÑÏ£ºÌïòÍ∏∞ ÏúÑÌï®Ïù¥Îã§.
  param : df (Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ)
  return : ÏÑûÏù∏ Îç∞Ïù¥ÌÑ∞
  """
  return df.sample(frac=1).reset_index(drop=True)

def split_target_data(df) :
  """
  X,y Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
  Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞ ÏÖãÏóê ÎßûÏ∂ò Ìï®Ïàò
  param : df (Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ)
  return : X, y
  """
  return df.iloc[:,range(8)], df['y']

def split_train_test_data(df, test_ratio) :
  """
  ÏÖîÌîå ÏóÜÏù¥ ÎπÑÏú®ÎåÄÎ°ú ÏïûÎí§Î°ú ÏûòÎùºÏÑú train, test Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
  param : df (Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ), test_ratio (ÌÖåÏä§Ìä∏ÏÖã ÎπÑÏú®)
  return : train, test
  """
  n = int(len(df)*(1-test_ratio))
  return df[:n], df[n:]

def normalize(df) :
  """
  minmax Ï†ïÍ∑úÌôî
  param : df (Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ)
  return : df (Ï†ïÍ∑úÌôîÎêú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ)
  """
  df = (df - df.min())/(df.max() - df.min())
  return df


def mini_batch(df, size) :
  """
  ÎØ∏ÎãàÎ∞∞Ïπò
  Ïù∏Îç±Ïä§ ÎÑòÎ≤ÑÎ°ú ÎØ∏ÎãàÎ∞∞ÏπòÎ•º Íµ¨ÌòÑ. ÏóêÌè¨ÌÅ¨ÎßàÎã§ ÎûúÎç§ Ï∂îÏ∂ú ÎêòÎèÑÎ°ù Ïù∏Îç±Ïä§Î•º ÏÑûÏñ¥ÏÑú Ï†úÎÑàÎ†àÏù¥ÌÑ∞Î°ú Îç∞Ïù¥ÌÑ∞ Î∞òÌôò
  param : df (Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ), size (ÎØ∏ÎãàÎ∞∞Ïπò ÌÅ¨Í∏∞)
  return : ÎØ∏ÎãàÎ∞∞Ïπò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ï†úÎÑàÎ†àÏù¥ÌÑ∞
  """
  index_list = list(df.index)
  random.shuffle(index_list)
  for i in range(0,len(index_list),size) :
    ind = index_list[i:i+size]
    yield split_target_data(df.loc[ind])

def sigmoid(x) :
  """
  ÏãúÍ∑∏Î™®Ïù¥Îìú
  param : x
  return : ÏãúÍ∑∏Î™®Ïù¥Îìú ÌÜµÍ≥ºÍ∞í
  """
  return 1/(1+np.e**(-x))

def forward_prop(input_data, weights, bias) :
  """
  Îç∞Ïù¥ÌÑ∞ ÌïòÎÇòÍ∞Ä Îì§Ïñ¥ÏôîÏùÑ Îïå Ïã†Í≤ΩÎßù ÌïúÏ∏µ Ïó∞ÏÇ∞
  param : input_data (Îã®Ïùº Îç∞Ïù¥ÌÑ∞), weights, bias
  return : Ïó∞ÏÇ∞ Í≤∞Í≥º np ÌñâÎ†¨
  """
  cal = input_data.dot(weights) + bias # ÏàúÏ†ÑÌåå ÌñâÎ†¨ Ïó∞ÏÇ∞
  sig_cal = sigmoid(cal) # ÌôúÏÑ±ÌôîÌï®Ïàò : ÏãúÍ∑∏Î™®Ïù¥Îìú
  return np.array(sig_cal)

def cal_lost(sig_cal, y) :
  """
  Î∞∞ÏπòÎ≥Ñ ÏÜêÏã§(Ïù¥ÏßÑ ÍµêÏ∞® ÏóîÌä∏Î°úÌîº)Í∞í Ïó∞ÏÇ∞
  param : sig_cal (Î∞∞Ïπò 1Í∞úÏùò Î™®Îç∏ ÏµúÏ¢Ö output ÌñâÎ†¨), y (Î∞∞Ïπò 1Í∞úÏùò ÌÉÄÍ≤üÍ∞í)
  return : Î∞∞ÏπòÏùò ÌèâÍ∑† Ïù¥ÏßÑ ÍµêÏ∞® ÏóîÌä∏Î°úÌîº
  """
  return -sum([y*np.log(p)+(1-y)*np.log(1-p) for p,y in zip(sig_cal, y)]) / len(y) # ÌèâÍ∑† Ïù¥ÏßÑ ÍµêÏ∞® ÏóîÌä∏Î°úÌîº Ïó∞ÏÇ∞

def accuracy(sig_cal, y, threshold = 0.5) :
  """
  Î∞∞ÏπòÎ≥Ñ ÌèâÍ∑† Ï†ïÌôïÎèÑ Ïó∞ÏÇ∞
  param : sig_cal (Î∞∞Ïπò 1Í∞úÏùò Î™®Îç∏ ÏµúÏ¢Ö output ÌñâÎ†¨), y (Î∞∞Ïπò 1Í∞úÏùò ÌÉÄÍ≤üÍ∞í), threshold (ÏûÑÍ≥ÑÍ∞í)
  return : Î∞∞ÏπòÏùò ÌèâÍ∑† Ï†ïÌôïÎèÑ
  """
  pred = [1 if i>=threshold else 0 for i in sig_cal]
  acc = [1 if p==y else 0 for p,y in zip(pred, y)]
  return sum(acc)/len(acc)

class model :
  """
  Î™®Îç∏ ÌÅ¥ÎûòÏä§
  param : input_dim (input Îç∞Ïù¥ÌÑ∞Ïùò Ï∞®Ïõê)
  """
  def __init__(self, input_dim) :
    """
    ÎÇ¥Ïû• Î≥ÄÏàò
    weights : Î†àÏù¥Ïñ¥Îì§ Í∞ÄÏ§ëÏπò
    bias : Î†àÏù¥Ïñ¥Îì§ Ìé∏Ìñ•
    dim : Î†àÏù¥Ïñ¥Îì§ Ï∞®Ïõê
    fin_dim : Ï∂úÎ†• Ï∞®Ïõê
    forward_prop_by_batch : Î∞∞ÏπòÎ≥Ñ ÏàúÏ†ÑÌåå Í≤∞Í≥ºÍ∞í
    loss_by_batch : Î∞∞ÏπòÎ≥Ñ loss Í∞í
    acc_by_batch : Î∞∞ÏπòÎ≥Ñ accuracy Í∞í
    """
    self.weights = [] # Î†àÏù¥Ïñ¥Îì§ Í∞ÄÏ§ëÏπò
    self.bias = [] # Î†àÏù¥Ïñ¥Îì§ Ìé∏Ìñ•
    self.dim = [input_dim] # Î†àÏù¥Ïñ¥Îì§ Ï∞®Ïõê
    self.fin_dim = self.dim[0] # ÎßàÏßÄÎßâ Ï∞®Ïõê

    self.forward_prop_by_batch = [] # Î∞∞ÏπòÎ≥Ñ ÏàúÏ†ÑÌåå Í≤∞Í≥ºÍ∞í
    self.loss_by_batch = [] # Î∞∞ÏπòÎ≥Ñ loss Í∞í
    self.acc_by_batch = [] # Î∞∞ÏπòÎ≥Ñ accuracy Í∞í

  def add_dense_layor(self, out_dim) :
    """
    Î™®Îç∏Ïùò Îß® ÎÅùÏóê out_dim Ï∞®ÏõêÏùò dense layor Î•º Ï∂îÍ∞ÄÌïúÎã§. Ï¥àÍ∏∞ÌôîÎêú ÌååÎùºÎØ∏ÌÑ∞ ÌñâÎ†¨ÏùÑ Ï†ÄÏû•ÌïúÎã§.
    param : out_dim (Ï∂îÍ∞ÄÌïòÎ†§Îäî dense layor Ïùò Ï∞®Ïõê)
    return : ÏóÜÏùå
    """
    in_dim = self.fin_dim # ÌòÑÏû¨ Î™®Îç∏Ïùò ÎßàÏßÄÎßâ Ï∞®ÏõêÏùÑ input Ï∞®ÏõêÏúºÎ°ú ÏÑ§Ï†ï
    w,b = initialization_parameter(in_dim, out_dim) # ÌååÎùºÎØ∏ÌÑ∞ Ï¥àÍ∏∞Ìôî
    self.weights.append(w) # ÌååÎùºÎØ∏ÌÑ∞ Ï†ÄÏû•
    self.bias.append(b)
    self.dim.append(out_dim) # Ï∞®ÏõêÍ∞í Ï†ÄÏû•
    self.fin_dim = out_dim # ÌòÑÏû¨ Î™®Îç∏Ïùò ÎßàÏßÄÎßâ Ï∞®ÏõêÍ∞í Ïû¨ÏÑ§Ï†ï

  def forward_propagation(self, batch) :
    """
    ÎØ∏ÎãàÎ∞∞ÏπòÎ•º Î∞õÏïÑÏÑú ÏàúÏ†ÑÌåå ÏßÑÌñâ, Î∞∞ÏπòÎ≥Ñ ÏàúÏ†ÑÌåå Í≤∞Í≥ºÏôÄ loss, accuracy Ï†ÄÏû•
    param : batch (ÎØ∏ÎãàÎ∞∞Ïπò Ï†úÎÑàÎ†àÏù¥ÌÑ∞)
    return : ÏóÜÏùå
    """
    # Ï¥àÍ∏∞Ìôî
    self.forward_prop_by_batch = []
    self.loss_by_batch = []
    self.acc_by_batch = []

    # Î∞∞ÏπòÎ≥Ñ ÏàúÏ†ÑÌåå Ïó∞ÏÇ∞ Î∞è loss, acc Ïó∞ÏÇ∞
    for X,y in batch :
      X = np.array(X)
      for w, b in zip(self.weights, self.bias) :
        X = forward_prop(X, w, b)
      self.forward_prop_by_batch.append(X)
      self.loss_by_batch.append(cal_lost(X, y))
      self.acc_by_batch.append(accuracy(X, y))




````

````python
def main() :
  # Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥ÌïòÏó¨ Ï†ÄÏû• Î∞è Î°úÎìú
  data_save(url = "https://drive.google.com/uc?export=download&id=1SCO0ZGL_EDGWc9Le0JFDw9eww86xk1xJ", file_name = "datafile.csv")
  df = data_load(file_path = "/content/datafile.csv")

  # Îç∞Ïù¥ÌÑ∞ ÏÑûÍ∏∞
  df = data_randmix(df)

  # train, test Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
  train, test = split_train_test_data(df, 0.2)

  # Î™®Îç∏ ÎßåÎì§Í∏∞
  M = model(8) # input dimension ÏùÑ ÌååÎùºÎØ∏ÌÑ∞Î°ú ÎÑòÍ∏∞Î©¥ÏÑú ÏãúÏûë = input layer
  M.add_dense_layor(5) # (ÎÖ∏Îìú 5Í∞ú) Î∞ÄÏßë Î†àÏù¥Ïñ¥ (+ ÏãúÍ∑∏Î™®Ïù¥Îìú) Ï∂îÍ∞Ä = hidden layer
  M.add_dense_layor(10) # (ÎÖ∏Îìú 10Í∞ú) Î∞ÄÏßë Î†àÏù¥Ïñ¥ (+ ÏãúÍ∑∏Î™®Ïù¥Îìú) Ï∂îÍ∞Ä = hidden layer
  M.add_dense_layor(1) # (ÎÖ∏Îìú 1Í∞ú) Î∞ÄÏßë Î†àÏù¥Ïñ¥ (+ ÏãúÍ∑∏Î™®Ïù¥Îìú) Ï∂îÍ∞Ä = output layer
  
  # train data Ïóê ÎåÄÌï¥ ÏóêÌè¨ÌÅ¨ 1ÌöåÎßå ÏßÑÌñâ
  for i in range(1,2) :

    # ÎØ∏ÎãàÎ∞∞Ïπò + X,y Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
    train_batch = mini_batch(train, 4)
    #test_batch = mini_batch(test, 4)

    # ÏàúÏ†ÑÌåå ÏßÑÌñâ
    M.forward_propagation(train_batch)
    
    # loss, accuracy Íµ¨ÌïòÍ∏∞
    loss = np.mean(M.loss_by_batch)
    accuracy = np.mean(M.acc_by_batch)
    
    print(f"[Epoch {i}] TrainData - Loss = {loss:.3f}, Accuracy = {accuracy:.3f}")
````

````python
main()
````

## Ïó≠Ï†ÑÌåå Ïù¥Î°†
Ïó≠Ï†ÑÌåå(Backpropagation)Îäî Î®ºÏ†Ä Í≥ÑÏÇ∞ Í≤∞Í≥ºÏôÄ Ï†ïÎãµÏùò Ïò§Ï∞®Î•º Íµ¨Ìï¥ Ïù¥ Ïò§Ï∞®Ïóê Í¥ÄÏó¨ÌïòÎäî Í∞íÎì§Ïùò Í∞ÄÏ¶ùÏπòÎ•º ÏàòÏ†ïÌïòÏó¨ Ïò§Ï∞®Í∞Ä ÏûëÏïÑÏßÄÎäî Î∞©Ìñ•ÏúºÎ°ú ÏùºÏ†ï ÌöüÏàòÎ•º Î∞òÎ≥µÌï¥ ÏàòÏ†ïÌïòÎäî Î∞©Î≤ïÏù¥Îã§. Ïò§Ï∞®Ïó≠Ï†ÑÌåå ÎòêÎäî Ïò§Î•òÏó≠Ï†ÑÌååÎùºÍ≥†ÎèÑ Î∂àÎ¶∞Îã§.



### ÌïôÏäµ Í≥ºÏ†ï
1. ÏûÑÏùòÏùò Ï¥àÍ∏∞ Í∞ÄÏ§ëÏπò($W$)Î•º Ï§Ä Îí§ Í≤∞Í≥º ($Y_{out}$)Î•º Í≥ÑÏÇ∞ÌïúÎã§.  
2. Í≥ÑÏÇ∞ Í≤∞Í≥ºÏôÄ Ïö∞Î¶¨Í∞Ä ÏõêÌïòÎäî Í∞í ÏÇ¨Ïù¥Ïùò Ïò§Ï∞®Î•º Íµ¨ÌïúÎã§.  
3. Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏùÑ Ïù¥Ïö©Ìï¥ Î∞îÎ°ú Ïïû Í∞ÄÏ§ëÏπòÎ•º Ïò§Ï∞®Í∞Ä ÏûëÏïÑÏßÄÎäî Î∞©Ìñ•ÏúºÎ°ú ÏóÖÎç∞Ïù¥Ìä∏ÌïúÎã§.  
4. ÏúÑ Í≥ºÏ†ïÏùÑ ÎçîÎäî Ïò§Ï∞®Í∞Ä Ï§ÑÏñ¥Îì§ÏßÄ ÏïäÏùÑ ÎïåÍπåÏßÄ Î∞òÎ≥µÌïúÎã§.  

### Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ï
Ïò§Ï∞®Í∞Ä ÏûëÏïÑÏßÄÎäî Î∞©Ìñ•ÏúºÎ°ú ÏóÖÎç∞Ïù¥Ìä∏ÌïúÎã§'Îäî ÏùòÎØ∏Îäî ÎØ∏Î∂Ñ Í∞íÏù¥ 0Ïóê Í∞ÄÍπåÏõåÏßÄÎäî Î∞©Ìñ•ÏúºÎ°ú ÎÇòÏïÑÍ∞ÑÎã§Îäî ÎßêÏù¥Îã§. Ï¶â, 'Í∏∞Ïö∏Í∏∞Í∞Ä 0Ïù¥ ÎêòÎäî Î∞©Ìñ•'ÏúºÎ°ú ÎÇòÏïÑÍ∞ÄÏïº ÌïòÎäîÎç∞, Ïù¥ ÎßêÏùÄ Í∞ÄÏ§ëÏπòÏóêÏÑú Í∏∞Ïö∏Í∏∞Î•º Î∫êÏùÑ Îïå Í∞ÄÏ§ëÏπòÏùò Î≥ÄÌôîÍ∞Ä Ï†ÑÌòÄ ÏóÜÎäî ÏÉÅÌÉúÎ•º ÎßêÌïúÎã§.
Ï†ÅÏ†àÌïú Í∞íÏúºÎ°ú Í∞íÏùÑ Í∞±Ïã†ÌïòÍ∏∞ ÏúÑÌï¥ ÌïôÏäµÎ•†(Œ±) Í∞íÏùÑ Ï°∞Ï†ïÌïúÎã§.
  $$W(t+1)=W(t)-Œ±\frac{\partial{Loss}}{\partial{W}}$$

Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏùÑ ÌÜµÌïú ÏµúÏÜåÍ∞í Ï∞æÍ∏∞Îäî local minimum Î¨∏Ï†úÏóê Îπ†Ïßà Ïàò ÏûàÎã§.

### Chain Rule
$$\frac{dy}{dx} = \frac{dy}{du}\cdot\frac{du}{dx}$$
Ï≤¥Ïù∏Î£∞ÏùÑ ÌÜµÌï¥ $\frac{\partial{Loss}}{\partial{W}}$ Í∞íÏùÑ Íµ¨Ìï† Ïàò ÏûàÎã§.

### Ïó≠Ï†ÑÌåå Î™®ÏãùÎèÑ

dense Î†àÏù¥Ïñ¥Ïùò Ïó≠Ï†ÑÌåå ÌïôÏäµÏùÄ ÌòÑÏû¨ Ï∏µÏùò input Í∞í, Îã§Ïùå Ï∏µÏóêÏÑú Î∞õÏïÑÏò¨ Ïàò ÏûàÎäî Ïó≠Ï†ÑÌåå Ïó∞ÏÇ∞Í∞íÏùò ÌñâÎ†¨Í≥±ÏúºÎ°ú ÌëúÌòÑÌï† Ïàò ÏûàÎã§.

## Ïó≠Ï†ÑÌåå Íµ¨ÌòÑ

````python
class dense :
  """
  dense layer class
  """
  def __init__(self, in_dim:int, out_dim:int) :
    """
    param
      in_dim (int) : input dimension
      out_dim (int) : output dimension
    """
    self.W , self.b = initialization_parameter(in_dim, out_dim)
    self.dW = None # Í∞ÄÏ§ëÏπò Î≥ÄÌôîÎüâ
    self.db = None # Ìé∏Ìñ• Î≥ÄÌôîÎüâ
    self.x = None # input data
  
  def forward(self, x:np.array) -> np.array :
    """
    dense Î†àÏù¥Ïñ¥Ïùò ÏàúÏ†ÑÌåå
    param
      x (np.array) : input data
    return
      out (np.array) : ÏàúÏ†ÑÌåå Í≤∞Í≥ºÍ∞í
    """
    self.x = x
    out = np.dot(x,self.W) + self.b
    return out

  def backward(self, dout:np.array, learning_rate:float) -> np.array : # dout : ÏßÅÏ†Ñ Ïó≠Ï†ÑÌåå ÏßÑÌñâÏóêÏÑú Î∞õÏïÑÏò® ÎØ∏Î∂ÑÍ∞í
    """
    dense Î†àÏù¥Ïñ¥Ïùò Ïó≠Ï†ÑÌåå
    param
      dout (np.array) : Ïù¥Ï†Ñ Ïó≠Ï†ÑÌååÏóêÏÑú Î∞õÏïÑÏò® Í∞í
      learning_rate (float) : ÌïôÏäµÎ•†
    return
      dx (np.array) : Îã§Ïùå Ïó≠Ï†ÑÌååÏóê ÎÑòÍ∏∏ Í∞í
    """
    dx = np.dot(dout,self.W.T) # Îã§Ïùå Ïó≠Ï†ÑÌååÏ∏µÏóêÏÑú ÏÇ¨Ïö©Ìï† ÎØ∏Î∂ÑÍ∞í
    self.dW = np.dot(self.x.T, dout) # Í∞ÄÏ§ëÏπò Î≥ÄÌôîÎüâ Í≥ÑÏÇ∞
    self.db = np.sum(dout, axis = 0) # Ìé∏Ìñ• Î≥ÄÌôîÎüâ Í≥ÑÏÇ∞
    self.W -= learning_rate * self.dW # Í∞ÄÏ§ëÏπò ÏóÖÎç∞Ïù¥Ìä∏
    self.b -= learning_rate * self.db # Ìé∏Ìñ• ÏóÖÎç∞Ïù¥Ìä∏

    return dx

class sigmoid :
  """
  sigmoid layer
  """
  def __init__(self) :
    self.out = None # ÏãúÍ∑∏Î™®Ïù¥Îìú ÌÜµÍ≥º Í≤∞Í≥ºÍ∞í

  def forward(self, x:np.array)->np.array :
    """
    ÏãúÍ∑∏Î™®Ïù¥Îìú ÏàúÏ†ÑÌåå
    param
      x (np.array) : input
    return
      out (np.array) : sigmoid(x)
    """
    out = 1/(1+np.exp(-x))
    self.out = out
    return out
  
  def backward(self, dout:np.array, _)->np.array :
    """
    ÏãúÍ∑∏Î™®Ïù¥Îìú Ïó≠Ï†ÑÌåå
    param
      dout (np.array) : Ïù¥Ï†Ñ Ïó≠Ï†ÑÌååÏóêÏÑú Î∞õÏïÑÏò® Í∞í
    return
      dx (np.array) : Îã§Ïùå Ïó≠Ï†ÑÌååÏóê ÎÑòÍ∏∏ Í∞í
    """
    dx = dout*(1.0-self.out)*self.out
    return dx

class relu :
  """
  relu layer
  """
  def __init__(self) :
    self.mask = None

  def forward(self, x:np.array)->np.array :
    """
    relu ÏàúÏ†ÑÌåå
    param
      x (np.array) : input
    return
      out (np.array) : relu(x)
    """
    self.mask = (x <= 0)
    out = x.copy()
    out[self.mask] = 0
    return out

  def backward(self, dout:np.array, _)->np.array :
    """
    relu Ïó≠Ï†ÑÌåå
    param
      dout (np.array) : Ïù¥Ï†Ñ Ïó≠Ï†ÑÌååÏóêÏÑú Î∞õÏïÑÏò® Í∞í
    return
      dx (np.array) : Îã§Ïùå Ïó≠Ï†ÑÌååÏóê ÎÑòÍ∏∏ Í∞í
    """
    dout[self.mask] = 0
    dx = dout
    return dx
  

class model_with_back_propagation :
  """
  Ïó≠Ï†ÑÌåå ÌïôÏäµ Íµ¨ÌòÑ Î™®Îç∏
  """
  def __init__(self, input_dim:int) :
    self.layers_info = [('input',input_dim)] # Î†àÏù¥Ïñ¥ Ï†ïÎ≥¥
    self.fin_dim = input_dim # ÎßàÏßÄÎßâ Ï∞®Ïõê
    self.layers = [] # Î†àÏù¥Ïñ¥ Î∞∞Ïó¥
    self.y = None # ÏòàÏ∏° Í≤∞Í≥ºÍ∞í
    self.t = None # Ïã§Ï†ú ÌÉÄÍ≤üÍ∞í
    self.loss = None # Î°úÏä§Í∞í
  
  def add_dense_layer(self, out_dim:int) :
    """
    Î™®Îç∏Ïóê dense layer Ï∂îÍ∞Ä
    param
      out_dim (int) : Î†àÏù¥Ïñ¥ Ï∞®Ïõê
    """
    self.layers.append(dense(self.fin_dim, out_dim))
    print(f"dense layer added : input dim = {self.fin_dim}, output dim = {out_dim}")
    self.layers_info.append(('dense',out_dim))
    self.fin_dim = out_dim

  def add_sigmoid_layer(self) :
    """
    Î™®Îç∏Ïóê sigmoid layer Ï∂îÍ∞Ä
    """
    self.layers.append(sigmoid())
    print(f"sigmoid layer added")
    self.layers_info.append(('sigmoid',self.fin_dim))

  def add_relu_layer(self) :
    """
    Î™®Îç∏Ïóê relu layer Ï∂îÍ∞Ä
    """
    self.layers.append(relu())
    print(f"relu layer added")
    self.layers_info.append(('relu',self.fin_dim))
  
  def forward_pass(self, x:np.array) :
    """
    Î™®Îç∏ ÏàúÏ†ÑÌåå ÏßÑÌñâ
    param
     x (np.array) : input data
    return
     x (np.array) : ÏàúÏ†ÑÌåå Í≤∞Í≥ºÍ∞í
    """
    for layer in self.layers :
      x = layer.forward(x)
    self.y = x
    return x
  
  def find_loss(self, t:np.array)->float :
    """
    Ïù¥ÏßÑ Î∞îÏù¥ÎÑàÎ¶¨ ÍµêÏ∞® ÏóîÌä∏Î°úÌîº loss Í∞í Í≥ÑÏÇ∞
    param
      t (np.array) : Ïã§Ï†ú ÌÉÄÍ≤üÍ∞í
    return
      self.loss (float) : ÌèâÍ∑† loss Í≥ÑÏÇ∞Í∞í
    """
    if not self.y.all() :
      print("Error : need forward_pass")
      return
    self.loss = np.squeeze(cal_lost(self.y, t))
    self.t = t
    return self.loss

  def acc(self, t:np.array)->np.array :
    """
    Ï†ïÌôïÎèÑ Í≥ÑÏÇ∞
    param
      t (np.array) : Ïã§Ï†ú ÌÉÄÍ≤üÍ∞í
    return
      ac : Ï†ïÌôïÎèÑ Ïó∞ÏÇ∞Í∞í
    """
    if not self.y.all() :
      print("Error : need forward_pass")
      return
    ac = accuracy(self.y,t)
    return ac

  def back_prop(self, t:np.array, learning_rate:float) :
    """
    Ïó≠Ï†ÑÌååÏôÄ ÌïôÏäµ
    param
      t (np.array) : Ïã§Ï†ú ÌÉÄÍ≤üÍ∞í
      learning_rate (float) : ÌïôÏäµÎ•†
    """
    t = np.array(t).reshape(-1,1)

    dout = -(t/self.y - (1-t)/(1-self.y)) # ÏÜêÏã§Ìï®Ïàò ÎØ∏Î∂Ñ Ïó≠Ï†ÑÌåå Í≥ÑÏÇ∞

    for layer in self.layers[::-1] : # Î†àÏù¥Ïñ¥ Ïó≠ÏàúÏúºÎ°ú Ïó≠Ï†ÑÌååÏôÄ ÌïôÏäµ ÏßÑÌñâ
      dout = layer.backward(dout, learning_rate)

````

````python
def main() :

  # Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥ÌïòÏó¨ Ï†ÄÏû• Î∞è Î°úÎìú
  data_save(url = "https://drive.google.com/uc?export=download&id=1SCO0ZGL_EDGWc9Le0JFDw9eww86xk1xJ", file_name = "datafile.csv")
  df = data_load(file_path = "/content/datafile.csv")

  # Îç∞Ïù¥ÌÑ∞ ÏÑûÍ∏∞
  df = data_randmix(df)

  # train, test Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
  train, test = split_train_test_data(df, 0.2)

  # split
  X_train, y_train = split_target_data(train)
  X_test, y_test = split_target_data(test)

  # Îç∞Ïù¥ÌÑ∞ Ï†ïÍ∑úÌôî
  X_train_norm = normalize(X_train)
  X_test_norm = normalize(X_test)
  train_norm = pd.concat([X_train_norm,y_train], axis=1)

  # Î™®Îç∏ ÎßåÎì§Í∏∞
  M = model_with_back_propagation(8)
  M.add_dense_layer(20)
  M.add_relu_layer()
  M.add_dense_layer(1)
  M.add_sigmoid_layer()

  print()

  # Ïó≠Ï†ÑÌåå Î∞è ÌïôÏäµ
  n = 1000 # ÏóêÌè¨ÌÅ¨ ÌöüÏàò
  batch_size = 4
  for i in range(1,n+1) : # ÏóêÌè¨ÌÅ¨ nÌöå
    # ÎØ∏ÎãàÎ∞∞Ïπò ÎßåÎì§Í∏∞
    train_batch = mini_batch(train_norm, batch_size)
    loss = 0
    acc = 0
    # Î∞∞ÏπòÎ≥Ñ ÌïôÏäµ
    for x,y in train_batch :
      M.forward_pass(x)
      if i%100 == 0 :
        loss += M.find_loss(y)
        acc += M.acc(y)
      else : M.find_loss(y)
      M.back_prop(y, 0.01)
    # Ï∂úÎ†•
    if i%100 == 0 :
      print(f"[Epoch {i}] TrainData - Loss = {loss/batch_size:.5f}, Accuracy = {acc/batch_size}")

  print()

  print("-----check test set-----")
  M.forward_pass(X_test_norm)
  print(f"test accuracy : {M.acc(y_test)}")
  pred = ((M.y >= 0.5).reshape(-1) == np.array(y_test))
  for i in pred :
    if i : print("correct")
    else : print("wrong")
````
